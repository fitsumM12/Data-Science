```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C']
values = [15, 24, 12]

# Create a column chart
sns.barplot(x=categories, y=values)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Column Chart')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 6, 8, 10]
y2 = [10, 8, 6, 10, 2]

# Create the first line plot
sns.barplot(x=x, y=y1, color="green")

# Create the second line plot on the secondary axis
plt.twinx()
sns.lineplot(x=x, y=y2, color='black')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y1-axis', color='blue')
plt.title('Dual Axis Chart')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 12]
values2 = [10, 18, 8]
values3 = [5, 12, 6]

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value1': values1, 'Value2': values2, 'Value3': values3})

# Reshape the DataFrame for stacked bar chart
df_stacked = df.melt('Category', var_name='Variable', value_name='Value')

# Create a stacked bar chart using Seaborn and Matplotlib
sns.barplot(x='Category', y='Value', hue='Variable', data=df_stacked, palette='muted')

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 12]
values2 = [10, 18, 8]

# Calculate the total values
total_values = [sum(values1), sum(values2)]

# Calculate the percentage values for each category
percent_values1 = [value / total_values[0] * 100 for value in values1]
percent_values2 = [value / total_values[1] * 100 for value in values2]

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value1': values1, 'Value2': values2, 'Percent1': percent_values1, 'Percent2': percent_values2})

# Sort the DataFrame by percent values
df_sorted = df.sort_values('Percent1', ascending=False)

# Calculate the cumulative sum of percent values
df_sorted['Cumulative1'] = df_sorted['Percent1'].cumsum()
df_sorted['Cumulative2'] = df_sorted['Percent2'].cumsum()

# Create the Mekko chart
fig, ax = plt.subplots()

# Plot the rectangles
ax.barh(df_sorted['Category'], df_sorted['Percent1'], label='Value1', color='skyblue', height=0.5)
ax.barh(df_sorted['Category'], df_sorted['Percent2'], left=df_sorted['Cumulative1'], label='Value2', color='lightgreen', height=0.5)

# Add labels and title
ax.set_xlabel('Percentage')
ax.set_ylabel('Category')
ax.set_title('Mekko Chart')

# Add legend
ax.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C']
sizes = [15, 30, 45]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Pie Chart')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, -5, 15, -7]

# Calculate cumulative sum
cumulative_values = [sum(values[:i+1]) for i in range(len(values))]

# Create a waterfall chart using Matplotlib
plt.bar(categories, values, color='gray')
plt.plot(categories, cumulative_values, marker='o', color='blue')

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Waterfall Chart')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [1000, 800, 600, 400]

# Calculate the percentage values
total = sum(values)
percentages = [(value/total)*100 for value in values]

# Create a funnel chart using Matplotlib
plt.barh(categories, percentages, color='skyblue')

# Reverse the y-axis to start from the top
plt.gca().invert_yaxis()

# Add labels and title
plt.xlabel('Percentage')
plt.ylabel('Category')
plt.title('Funnel Chart')

# Display the chart
plt.show()
```
